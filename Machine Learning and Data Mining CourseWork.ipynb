Applying PCA

```python
from sklearn.model_selection import train_test_split

X = features_plus_classes.iloc[:,0:2000]  #independent columns
y = features_plus_classes.iloc[:,-1]   
# test_size: what proportion of original data is used for test set
train_img, test_img, train_lbl, test_lbl = train_test_split( X, y, test_size=1/7.0, random_state=0)
```

PCA is effected by scale so you need to scale the features in the data before applying PCA.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit on training set only.
scaler.fit(train_img)

# Apply transform to both the training set and the test set.
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
```

Import and Apply PCA

Notice the code below has .95 for the number of components parameter. It means that scikit-learn choose the minimum number of principal components such that 95% of the variance is retained.

```python
from sklearn.decomposition import PCA
# Make an instance of the Model
pca = PCA(.95)
```

```python
pca.fit(train_img)
```

```python
train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
from sklearn.linear_model import LogisticRegression
# all parameters not specified are set to their defaults
# default solver is incredibly slow which is why it was changed to 'lbfgs'
logisticRegr = LogisticRegression(solver = 'lbfgs')
logisticRegr.fit(train_img, train_lbl)
# Predict for One Observation (image)
logisticRegr.predict(test_img[0].reshape(1,-1))
# Predict for One Observation (image)
logisticRegr.predict(test_img[600:700])
logisticRegr.score(test_img, test_lbl)
```

```python
# Naive bayes gausian on the raw data

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X = features_plus_classes.iloc[:,0:2000]  #independent columns
y = features_plus_classes.iloc[:,-1]   
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=1/7.0, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
#print("Number of mislabeled points out of a total %d points : %d"(X_test.shape[0], (y_test != y_pred).sum()))
print("total:",X_test.shape[0],"Misclassified:",(y_test != y_pred).sum())
```
